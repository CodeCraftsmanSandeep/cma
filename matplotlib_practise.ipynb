```python
# practising bar plot
# simulating dice
import random
import matplotlib.pyplot as plt

barwidth = 0.25
fig = plt.subplots(figsize =(12, 8)) 

li = [1, 2, 3, 4, 5, 6]
freq1 = [0 for i in range(6)]
for i in range(1000):
    freq1[random.choice(li)-1] += 1

freq2 = [0 for i in range(6)]
for i in range(1000):
    freq2[random.choice(li)-1] += 1

br1 = [i for i in range(1, 7)]
br2 = [i + barwidth for i in range(1, 7)]
plt.bar(br1, freq1, width = barwidth, label = "simulation 1")
plt.bar(br2, freq2, width = barwidth, label = "simulation 2")
plt.xticks([i + barwidth/2 for i in range(1, 7)], li)
plt.legend()
plt.title('Multiple dice simulations')

plt.show()
```
