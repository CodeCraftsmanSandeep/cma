<h4> Square and circle<h4>

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the radius of the circle
radius = 1

# Generate angles from 0 to 2*pi
theta = np.linspace(0, 2*np.pi, 100)

# Calculate x and y coordinates for each angle
x = radius * np.cos(theta)
y = radius * np.sin(theta)

# Plot the circle
plt.figure()
plt.plot(x, y)

# Set equal aspect ratio to make the circle appear circular
plt.axis('equal')

# Add title and labels
plt.title('Circle Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(linestyle = '--')

# Display the plot
plt.show()
```
